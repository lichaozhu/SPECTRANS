```python
text_file = open("test-anno", "r")
 
#read whole file to a string
data = text_file.read()
 
#close file
text_file.close()
 
len(data)
nlp.max_length = len(data) + 100
```

```python
nlp = spacy.load("en_core_web_sm", exclude=["parser"])
nlp.enable_pipe("senter")
doc = nlp(data)
l = []
l_json = []

for sent in doc.sents:
    
    print(sent.text)
    for token in nlp(sent.text):
        l.append([token.text, token.lemma_, token.pos_, token.tag_, token.dep_,
            token.shape_, token.is_alpha, token.is_stop,sp.encode_as_pieces(str(token))])
        
        l_json.append([token.text, token.pos_])
    print("-------------------------------------------")
```

```python
with open('list.json', 'w') as F:
    # Use the json dumps method to write the list to disk
    F.write(json.dumps(l_json))
```

```python
col_names = ["TEXT","LEMMA","UPOS","XPOS","DEP","SHAPE","ALPHA","STOP","SentencePiece"]
!pip install tabulate

from tabulate import tabulate
```

```python
print(tabulate(l, headers=col_names))
```
