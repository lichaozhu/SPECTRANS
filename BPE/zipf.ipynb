```python
from nltk.tokenize import word_tokenize
corpus_path = "/home/nmt_shared/training_data/bpe/sentencepiece/europarl/europarl-train-sen.fr"
corpus = open (corpus_path,encoding="utf-8")
dict_word_corpus = dict()
for line in corpus:
    # Remove the leading spaces and newline character
    line = line.strip()

    # Convert the characters in line to
    # lowercase to avoid case mismatch
    line = line.lower()


    # Split the line into words

    corpus_words = word_tokenize(line)
    for word in corpus_words:
        if ort.check(word):
            # Check if the word is already in dictionary
            if word in dict_word_corpus:
                # Increment count of word by 1
                dict_word_corpus[word] = dict_word_corpus[word] + 1
            else:
                # Add the word to dictionary with count 1
                dict_word_corpus[word] = 1
```

```python
for key, value in dict_word_corpus.items():
    print(key, ' : ', value) 
```

```python
del dict_word_corpus['.']
```

```python
max_frequence = max(dict_word_corpus.keys(), key=(lambda k: dict_word_corpus[k]))
print("la frequence max d'un mot dans le corpus est :",dict_word_corpus[max_frequence])
print(max_frequence)
```

```python
#dict pour les mots presents dans le vocabulaire sans etre sous-tokenizer
non_token = dict()
for mot_vocabulaire in liste_vocabulaire:
    if mot_vocabulaire in dict_word_corpus.keys():
        non_token[mot_vocabulaire] = dict_word_corpus[mot_vocabulaire]
```

```python
print("The length of the dictionary is {}".format(len(non_token)))
#il ya donc 23100 token present dans le vocabulaire qui ne sont pas sous tokenizer
```

```python
non_token_sorted = sorted(non_token.items(), key=lambda x: x[1], reverse=True)
```

```python
print(non_token_sorted)
```

```python
import pandas as pd
```

```python
df = pd.DataFrame(non_token_sorted)
```

```python
print(df)
```

```python
df =df.rename(columns={0:'word', 1:'count'})
```

```python
df.index = df.index + 1

df.head()
```

```python
df['proportion'] = df['count']/df['count'].sum()
df.head()
```

```python
vocab_size, _ = df.shape
n_tokens = df['count'].sum()

print(f'Vocabulary size: {vocab_size:>8,}\nNumber of tokens: {n_tokens:,}')
```

```python
import numpy as np
def classic_zipf(N, k, s=1):
    return (1/k**s)/(np.sum(1/(np.arange(1, N+1)**s)))
```

```python
N = n_tokens
```

```python
%%time
vectorize_classic_zipf = np.vectorize(lambda x: classic_zipf(N, x))

df['predicted_proportion'] = vectorize_classic_zipf(df.index.values)
```
